```python
# Ordinal Encoding / Custom Mapping
import category_encoders as ce

yesNoCols = ['default', 'housing', 'loan', 'y']

mapping = [
    *list(map(lambda x: {'col': x, 'mapping': {None: 0, 'no': 0, 'yes': 1}}, yesNoCols)),
    {
        'col': 'month',
        'mapping': {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9,
                    'oct': 10, 'nov': 11, 'dec': 12}
    },
]

print(mapping)

data = (ce.OrdinalEncoder(cols=yesNoCols, mapping=mapping).fit_transform(data))

data
```

```python
# One-Hot (Dummy) Encoding

# columns_to_encode = ['job', 'marital', 'education', 'contact', 'poutcome']
# 
# for col in columns_to_encode:
#     dummies = pd.get_dummies(data[col], prefix=col, dtype=int, drop_first=True)
#     data = pd.concat([data, dummies], axis=1)
#     data.drop(col, axis=1, inplace=True)
# 
# data

import category_encoders as ce

data = ce.OneHotEncoder(
    cols=['job', 'marital', 'education', 'contact', 'poutcome'],
    use_cat_names=True,
    drop_invariant=True,
    handle_missing='ignore'
).fit_transform(data)
data
```

**pdays** feature represents number of days after the client was contacted (-1 means client was not contacted)

```python
data['was_contacted'] = data['pdays'].apply(lambda x: 1 if x > -1 else 0)
data.drop('pdays', axis=1, inplace=True)
data
```

```python
data.to_csv('../data/output/bank+marketing-bank_encoded.csv', index=False)
```
